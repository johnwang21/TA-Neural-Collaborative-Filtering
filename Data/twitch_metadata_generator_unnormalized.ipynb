```python
#requires original data set (100k_a.csv), train and test sets
import pandas as pd
import numpy as np

df_train = pd.read_csv('100k_a.csv')
df_train = df_train.rename(columns={"1": "UserID", "33842865744": "StreamID", "mithrain": "ChannelID", "154": "Start", "156": "Stop"})
df_train.head()
```

```python
frequency = df_train.groupby(["UserID", "ChannelID","StreamID"]).count()
frequency
#you want ChannelID not StreamID
```

```python
frequency = df_train.groupby(["UserID", "ChannelID"]).count()
frequency
```

```python
frequency = frequency.reset_index()
frequency
```

```python
frequency = frequency.iloc[:,[0,1,2]]
frequency
```

```python
frequency = frequency.rename(columns={"StreamID": "Frequency"})
frequency
```

```python
#now we try to see how many unique channels a user watches
cycle = df_train.groupby(["UserID", "ChannelID"]).count().reset_index()
cycle
#this groups multiple watches
```

```python
cycle = cycle.iloc[:,[0,1]]
User_Deg = cycle.groupby(["UserID"]).count()
User_Deg
```

```python
User_Deg = User_Deg.reset_index()
User_Deg
```

```python
User_Deg = User_Deg.iloc[:,[0,1]]
User_Deg = User_Deg.rename(columns={"ChannelID": "User_Deg"})
User_Deg
```

```python
#gets channel degree
df_train.groupby(["ChannelID", "UserID"]).count()
#removes repeated watches
```

```python
cycle2 = df_train.groupby(["ChannelID", "UserID"]).count().reset_index()
cycle2
```

```python
channel_links = cycle2.groupby(["ChannelID"]).count().reset_index()
channel_links
```

```python
Channel_Deg = channel_links.iloc[:,[0,1]]
Channel_Deg = Channel_Deg.rename(columns={"UserID": "Channel_Deg"})
Channel_Deg
```

```python
real_train = pd.read_csv('train.csv')
real_train.head()
```
